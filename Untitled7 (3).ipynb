```python
# 1
import mysql.connector
from mysql.connector import errorcode
import pandas as pd

#Importing Coal Dataset
Coal_df = pd.read_csv("Coal_Dataset.csv")
Coal_df.head()

#Dropping Code column from Coal and Biomass dataset as it wont be used for further queries as we have countries and continents names the Code is not required.
Coal_df = Coal_df.drop(['Code'], axis=1)

Coal_df.head()
```

```python
# 2
import numpy as np
Biomass_df = pd.read_csv("Biomass_Dataset.csv")
#Dropping Code column from Coal and Biomass dataset as it wont be used for further queries as we have countries and continents names the Code is not required.
Biomass_df = Biomass_df.drop(['Code'], axis=1)

#Filling all nan columns with value 0.
Biomass_df = Biomass_df.replace(np.nan, 0)
Biomass_df.head(75)
```

```python
# 3
import mysql.connector
from mysql.connector import errorcode
import warnings
warnings.filterwarnings('ignore')

myconnection = mysql.connector.connect(user='devanshi_patel1', password='7NTI75GCO',host='datasciencedb.ucalgary.ca', database='devanshi_patel1')

# CREATE TABLE STATEMENT for Electricity generated by Coal
create_statement = '''create table Coal_electricity (
    Country varchar(40) NOT NULL,
    Year int,
    TWh int
    );'''

# now we'll create a cursor and run our create statement
create_cursor = myconnection.cursor()
try:
    create_cursor.execute(create_statement)
except mysql.connector.Error as err:
    if err.errno == errorcode.ER_TABLE_EXISTS_ERROR:
        print("Ooops! We already have that table")
    else:
        print("Error Occured!")
else:
    print("table created successfully!")

create_cursor.execute(create_statement)
create_cursor.close()
```

```python
# 7
#Utility function for printing output depending on the query.
def print_queries(read_cursor):
    for (library_value) in read_cursor:
        print(library_value)
```

```python
# 8
#For Coal dataset, Find all the continents and group of countries except individual country names. For Example : World, Asia, Europe are not countires they are either continents or Group of countries.
#Using regex to find out pattern such as Asia% means anything starting with Asia followed by any string.
read_cursor = myconnection.cursor(buffered=True, named_tuple=True)

query_string_only_continents_group_countries = (
"SELECT SUM(TWh) as TWh_Sum, Country FROM Coal_electricity GROUP BY Country HAVING Country like 'Asia%' OR Country LIKE 'Europe%' OR Country LIKE 'South%America%' OR Country LIKE 'North%America%' OR Country LIKE '%(BP)' OR Country LIKE '%countries' OR Country LIKE '%(Ember)' OR Country LIKE 'Oceania' OR Country LIKE 'World' ORDER BY TWh_Sum DESC;")

read_cursor.execute(query_string_only_continents_group_countries)

#Converting output into dataframe for visualization.
columns = [desc[0] for desc in read_cursor.description]
data_df = pd.DataFrame(read_cursor.fetchall(), columns=columns)
print_queries(read_cursor);
read_cursor.close()

data_df.head(10)

import plotly.express as px
fig = px.bar(x=data_df["Country"], y=data_df["TWh_Sum"])
fig.show()

read_cursor.close()
```

```python
# 9
#For Coal dataset, Find all the individuaL country names. For Example : China, India, US,Canada and arrange them from highest to lowest production of electricity and group by country. 
#Using the NOT keyword and same pattern with AND instead of OR we get the required result of all individual countries which are not continents and group of countries.

read_cursor = myconnection.cursor(buffered=True, named_tuple=True)

query_string_only_countires = (
"SELECT SUM(TWh) as TWh_Sum, Country FROM Coal_electricity GROUP BY Country HAVING Country NOT like 'Asia%' AND Country NOT LIKE 'Europe%' AND Country NOT LIKE 'South%America%' AND Country NOT LIKE 'North%America%' AND Country NOT LIKE '%(BP)' AND country NOT LIKE '%countries' AND Country NOT LIKE '%(Ember)' AND Country NOT LIKE 'Oceania' AND Country NOT LIKE 'World' ORDER BY TWh_Sum DESC;")
#Find all the individual country names.
read_cursor.execute(query_string_only_countires)

#Converting output into dataframe for visualization.
columns = [desc[0] for desc in read_cursor.description]
data_df = pd.DataFrame(read_cursor.fetchall(), columns=columns)
print_queries(read_cursor);
read_cursor.close()

data_df.head(10)

import plotly.express as px
fig = px.bar(x=data_df["Country"], y=data_df["TWh_Sum"])
fig.show()

read_cursor.close()
```
